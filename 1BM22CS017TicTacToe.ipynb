```python
import random

def check(row, col):
    return arr[row][col] == '_'

def check2():
    return any('_' in row for row in arr)

def print_board():
    for row in arr:
        print(" | ".join(row))
        print("-" * 9)

def check_win(player):
    for i in range(3):
        if all(arr[i][j] == player for j in range(3)):
            return True
        if all(arr[j][i] == player for j in range(3)):
            return True
    if arr[0][0] == player and arr[1][1] == player and arr[2][2] == player:
        return True
    if arr[0][2] == player and arr[1][1] == player and arr[2][0] == player:
        return True
    return False

def computer_move():
    # Check for winning move
    for row in range(3):
        for col in range(3):
            if check(row, col):
                arr[row][col] = 'O'
                if check_win('O'):
                    return  # Block winning move
                arr[row][col] = '_'  # Reset the spot

    # Check for player's winning move and block it
    for row in range(3):
        for col in range(3):
            if check(row, col):
                arr[row][col] = 'X'
                if check_win('X'):
                    arr[row][col] = 'O'  # Block player's winning move
                    return
                arr[row][col] = '_'  # Reset the spot

    # If no winning move to block, pick a random spot
    while True:
        row, col = random.randint(0, 2), random.randint(0, 2)
        if check(row, col):
            arr[row][col] = 'O'
            return

arr = [['_', '_', '_'],
       ['_', '_', '_'],
       ['_', '_', '_']]
print_board()

while check2():
    row = int(input("Enter row (0-2): "))
    col = int(input("Enter column (0-2): "))
    while not check(row, col):
        print("Place already occupied, try again.")
        row = int(input("Enter row (0-2): "))
        col = int(input("Enter column (0-2): "))
    arr[row][col] = 'X'
    print_board()
    if check_win('X'):
        print("Congratulations! You win!")
        break
    if not check2():
        print("It's a draw!")
        break

    print("Computer's turn...")
    computer_move()
    print_board()
    if check_win('O'):
        print("Computer wins! Better luck next time.")
        break
    if not check2():
        print("It's a draw!")
        break

print("Aditya Dinesh Netrakar")
print("USN: 1BM22CS017")
```
